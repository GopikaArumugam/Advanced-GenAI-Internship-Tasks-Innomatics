```python
#1.validate mobile recharge
amt=int(input("Recharge"))
plans=[199,299,399,599]
while not(amt>=50 and amt in plans):
  print("Retry")
  amt=int(input())
print("Recharge Sucessful")
```

```python
#2.Check product stock
d={"Apple":50,"Orange":25,"Grapes":14}
for k,v in d.items():
  if v>=15:
    print(f"{k} : STOCK OK")
  else:
    print(f"{k} : REORDER ALERT")
```

```python
#3.Student Result Processing System
marks=eval(input())
avg=sum(marks)//len(marks)
if avg>=50:
  print("Pass")
else:
  print("Fail")
```

```python
#4.Cab fare estimator with retry option
def calc_far(dis,peak):
  base=50
  fare=(dis*12)+base
  if peak=="yes":
    fare+=fare*0.25
  return fare
while True:
  dis=int(input())
  peak=input("Is it peak hour?(yes/no)")
  print(calc_far(dis,peak))
  retry=input("Do you want to calculate for another fare?")
  if retry=="no":
    break
print("Thank you for using it")
```

```python
#5.Employee Attendance eligibility
l=input("Enter attendance (P/A) separated by space: ").split()
p=l.count("P")
avg=p/len(l)
print("Eligible") if avg>=0.75 else print("Not Eligible")
```

```python
#6.Password Strength Checker
s=input()
spl=False
dig=False
for ch in s:
  if ch.isdigit():
    dig=1
  if ch in "!@#$%^&*":
    spl=1
if len(s)>=8 and spl and dig:
  print("Valid")
else:
  print("Invalid")
```
