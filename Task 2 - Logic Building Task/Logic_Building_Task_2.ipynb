```python
#1.Unique Words in a Sentence
s=input().split()
wrds=set(s)
for w in s:
  if s.count(w)==1:
    wrds.add(w)
print(wrds,len(wrds))
```

```python
#2.Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000}
max_employee=max(employees, key=employees.get)
print("Highest Salary:",max_employee,"-",employees[max_employee])
```

```python
#3.Maximum and Minimum
numbers=[45, 22, 89, 10, 66]
numbers.sort()
print("Maximum:",numbers[-1])
print("Minimum:",numbers[0])
```

```python
#4.Count Products Above a Price Threshold
prices=[450,1200,899,1500,300]
cnt=0
for p in prices:
  if p>=1000:
    cnt+=1
print("Products above 1000:",cnt)
```

```python
#5.Calculate Attendance Percentage
attendance=["P","P","A","P","P"]
cnt=0
for ch in attendance:
  if ch=="P":
    cnt+=1
print("Attendance Percentage:",(cnt/len(attendance))*100)
```

```python
#6.Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
print(set(phone_numbers))
```

```python
#7.Count Character Frequency
s=input()
d={}
for ch in s:
  if ch in d:
    d[ch]+=1
  else:
    d[ch]=1
print(d)
```

```python
#9.Check if a Key Exists in a Dictionary
employee={
    "id": 101,
    "name": "John",
    "department": "HR"
}
key="name"
if key in employee:
    print("Employee exists")
else:
    print("Employee does not exist")
```
