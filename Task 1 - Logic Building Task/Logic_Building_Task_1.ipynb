```python
#Pass / Fail Analyzer
marks=[45,78,90,33,60]
ps=0
for m in marks:
  if m>=50:
    ps+=1
print("Pass:",ps)
print("Fail:",len(marks)-ps)
```

```python
#Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
res=[]
for w in names:
  res.append(w.strip().lower())
print(res)
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
msg_len=[]
flag=0
for w in messages:
  n=len(w)
  if n>10:
    flag+=1
  msg_len.append(n)
print(msg_len)
print("Flagged Count:",flag)
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
cnt=0
for w in logs:
  if w=="ERROR":
    cnt+=1
print(cnt)
```
