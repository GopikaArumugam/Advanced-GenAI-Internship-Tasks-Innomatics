```python
#1.Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85}
val=max(employees.values())
for k,v in employees.items():
    if v==val:
        print(k)
```

```python
#2.Search Query Keyword Analysis
s=input().lower().split()
f={}
for w in s:
  if w in f:
    f[w]+=1
  else:
    f[w]=1
for k,v in f.items():
  if v>=2:
    print(f"{k}:{v}")
```

```python
#3.Sensor Data Validation
readings=eval(input())
lst=[]
for i in range(len(readings)):
  if readings[i]%2==0:
    lst.append((i,readings[i]))
print(lst)
```

```python
#4.Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
total=len(emails)

for domain, count in domain_count.items():
    percentage=(count / total) * 100
    print(f"{domain}:{int(percentage)}%")
```

```python
#5.Sales Spike Detection
l=eval(input())
avg=sum(l)/len(l)
val=(0.3*avg)+avg
for i in range(len(l)):
  if l[i]>val:
    print("Day",i+1,":",l[i])
```

```python
#6.Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
duplicate_counts = {}
for user in user_ids:
    if user_ids.count(user) > 1:
        duplicate_counts[user] = user_ids.count(user)

print("Duplicate User IDs and their counts:")
for user, count in duplicate_counts.items():
    print(user, "->", count, "times")
```
